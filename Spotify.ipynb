```python
#makes the "release_date" column as the new index of the DataFrame 
#df_tracks and changes its format to work with dates.
df_tracks.set_index("release_date",inplace=True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
#converting duration in ms to just secs 
df_tracks["duration"]= df_tracks["duration_ms"].apply(lambda x:(x/1000))
df_tracks.drop("duration_ms",inplace=True,axis=1)
#significance of axis=1 means you are dropping column
#DataFrames have only two axes: axis=0 for rows and axis=1 for columns.
#So, when working with DataFrames, axis=0 represents operations along the rows, 
#and axis=1 represents operations along the columns.
```

```python
df_tracks.duration.head()
```

```python
#co relation df , heatmap
corr_df=df_tracks.drop(["key","mode","explicit"],axis=1).corr(method="pearson",numeric_only=True)
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True , fmt=".1f",vmin=-1,vmax=1 , center = 0, cmap="inferno",linewidths=1 , linecolor="Black")
heatmap.set_title("Correlation HeatMap between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
year=df_tracks.dates.dt.year
```

```python
year
```

```python
#observe the trend of increasing number of songs through the years
#histogram
sns.displot(year,discrete=True,aspect=2,height=5,kind="hist").set(title="Number of songs per year")
```

```python
#year vs Duration
total_dr = df_tracks.duration
fig_dims=(18,7)
fig,ax = plt.subplots(figsize = fig_dims)
fid=sns.barplot(x=year , y = total_dr , ax=ax , errwidth= False ).set(title="Year vs Duration")
plt.xticks(rotation=90)
```

```python
#lineplot to get the average duration of the songs over the years 
#same as bar plot but now in terms of a line
total_dr=df_tracks.duration
sns.set_style(style="whitegrid")
fig_dims=(10,5)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=year,y=total_dr,ax=ax).set(title="Year vs Duration")
plt.xticks(rotation=60)
```

```python
df_genre=pd.read_csv("SpotifyFeatures.csv")
```

```python
df_genre.head()
```

```python
#bar plot for duration of songs in different genres
plt.title("Duration of songs in Different Genres")
sns.color_palette("rocket",as_cmap=True)
sns.barplot(y='genre',x='duration_ms',data=df_genre)
plt.xlabel("Duration in milliseconds")
plt.ylabel("Genres")
```

```python
#top 5 genres by popularity 
sns.set_style(style="darkgrid")
plt.figure(figsize=(10,5))
famous=df_genre.sort_values("popularity",ascending=False).head(10)
sns.barplot(y='genre',x='popularity',data=famous).set(title="Top 5 Genres by Popullarity")
```
